##3.1.2 $\dot{x}=r-\cosh{x}$

Existe un punto estable e inestable cuando $r\geq1$. Para anlizar la curva de los puntos fijos, declaramos $\dot{x}=r-\cosh{(x)}=0$. Reacomodando los términos, se tiene que $x_{1,2}=\pm \cosh^{-1}{(r)}$. Como el $\cosh{(x)}$ no puede ser menor a $1$, $\cosh^{-1}{(r)}$ debe tener un argumento $r\geq 1$. Luego, $x_2=-\cosh^{-1}{(r)}$ es el punto inestable.

```python
x=np.linspace(-10,10,100)
r=0
x_dot=r-np.cosh(x)
fig = plt.figure()


plt.axis([-3, 3, -8, 0])
plt.plot(x,x_dot)
plt.grid()
#Para r=0 no hay raíces
```

```python
x=np.linspace(-10,10,100)
r=1
x_dot=r-np.cosh(x)
fig = plt.figure()


plt.axis([-3, 3, -6, 1])
plt.plot(x,x_dot)
plt.grid()

x=sp.symbols('x')
x_dot1=r-sp.cosh(x)
roots=sp.solve(x_dot1)
print(roots)
for j in range(len(roots)):
  plt.plot(roots[j],0,'ro')

#Luego hay un punto semi-estable en 1
```

```python
x=np.linspace(-10,10,100)
r=2
x_dot=r-np.cosh(x)
fig = plt.figure()


plt.axis([-3, 3, -6, 1])
plt.plot(x,x_dot)
plt.grid()

x=sp.symbols('x')
x_dot1=r-sp.cosh(x)
roots=sp.solve(x_dot1)
print(roots)
for j in range(len(roots)):
  plt.plot(roots[j],0,'ro')

#Luego hay un punto inestable y un punto inestable en el resultado a continuación:
```

```python
def fun (t,x,r,b):
  return r-np.cosh(x)


ti=0; tf=100; dt=100
tspam=[ti, tf]

for r in np.arange(-5,5,0.2):
  for x0 in np.arange(-5,5,0.7):
    sol=solve_ivp(fun,tspam,[x0],args=(r,b))
    plt.scatter(r, sol.y[0][-1], color='darkgreen', marker='.')

tspam=[ti, -tf]

for r in np.arange(-5,5,0.2):
  for x0 in np.arange(-5,5,0.7):
    sol=solve_ivp(fun,tspam,[x0],args=(r,b))
    plt.scatter(r, sol.y[0][-1], color='red', marker='.')


plt.axis([0, 3, -3, 3])
plt.grid()
```

##5.1.4 Escribir el siguiente sistema en forma matricial: $\dot{x}=3x-2y,\;\;$ $\dot{y}=2y-x$

```python
x,y = np.meshgrid(np.linspace(-20,20,10),np.linspace(-20,20,10))
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_xticks([])
ax.set_yticks([])
ax.spines['left'].set_position(('data',0.0))
ax.spines['bottom'].set_position(('data',0.0))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.set_xlabel("x",loc='right')
ax.set_ylabel("y", loc='top')

u = 3*x-2*y
v = 2*y-x

plt.quiver(x,y,u,v)
plt.show()
```
